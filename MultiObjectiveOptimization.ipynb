```python
import numpy as np
from pymoo.core.problem import Problem


class algoInput(Problem):

    def _evaluate(self, x, out, *args, **kwargs):
        res = []
        constr = []
        for i in x:
          
          in1 = scaler.transform([i])
          in2 = enerScaler.transform([i])
          print(in1,in2)
          
          res.append([modelGHG.predict(in1),101-modelEner.predict(in2)])
          #print(res[-1])
          constr.append((i[0] + i[1] + i[2]) - (i[3]))

        out["F"] = np.array(res)
        out['G'] = np.array(constr)
```
